```python
def sinplot(flip=1):
  x=np.linspace(0,15,100)
  for i in range(1,5):
    plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
sns.set_style('darkgrid')
sns.set_palette('deep')
sinplot()
plt.show()
```

#data frame-Titanic

```python
df1= sns.load_dataset('titanic')
```

```python
df1
```

```python
#draw a histogram with the age column
sns.distplot(df1['age'])
plt.show()
```

```python
#draw a barplot using pclass and fare
sns.barplot(x='pclass',y='fare',data=df1,color='black')
plt.show()
```

```python
#draw a scatter plot using age and fare
sns.scatterplot(x='age',y='fare',data=df1,color='purple')
plt.show()
```

```python
#draw a violine plot using fare
sns.violinplot(x='fare',data=df1)
plt.show()
```

```python
#draw a pair plot
sns.pairplot(df1)
plt.show()
```

```python
#draw a facet grid with pclass,sex and fare
g=sns.FacetGrid(df1,col='pclass',row='sex')
g.map(plt.hist,'fare',color='purple')
plt.show()
```

```python
#plot a heatmap
sns.heatmap(df1.select_dtypes(include=np.number).corr(),annot=True)
plt.show()
```

```python
#plot a cluster map
sns.clustermap(df1.select_dtypes(include=np.number).corr(),annot=True)
plt.show()
```

```python
#draw a join plot using age and fare
sns.jointplot(x='age',y='fare',data=df1,color='purple')
plt.show()
```
